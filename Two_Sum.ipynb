brute force

```python
from typing import List

class Solution:
  def twoSum(self, nums: List[int], target: int ) -> List[int]:

    for i in range(len(nums)):
            for j in range(i + 1, len(nums)):
                if nums[i] + nums[j] == target:
                  return[i, j]
    return None

k = Solution()
print(k.twoSum([1,2,3,4],5))
print(k.twoSum([3,4,5],19))
```

```python
from typing import List

class Solution:
  def twoSum(self, nums: List[int], target: int) -> List[int]:
      prevMap = {} # val -> index

      for i, n in enumerate(nums):
              diff = target - n
              if diff in prevMap:
                   return [prevMap[diff], i]
              prevMap[n] = i
      return

k = Solution()
print(k.twoSum([1,2,3,4],7))
```
